# Задание 2. 
Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

Рассмотрим показатель Pulse. при расчете около 10% удалилось, что показало отклонение среднее арифметическое изменилось на 1,65, это примерно на 2,5% от изначальных показателей. Думаю, что это отклонение при удалении выбросов допустимое при max = 184.0 и min = 30.0

```python
#возьмем 
print(df_new.pulse.mean())
print(df_new[(df_new.pulse > 40) & (df_new.pulse < 140)].pulse.mean())
```

```python
print(df_new.pulse.median())
print(df_new[(df_new.pulse > 40) & (df_new.pulse < 140)].pulse.median())
```

```python
print(df_new.pulse.round().mode()[0])
```

```python
q1 = df_new.pulse.quantile(0.25)
q3 = df_new.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) #нижняя граница не выбросов
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df_new[df_new.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('age')
remove_outliers
```

```python
print(remove_outliers.pulse.mean())
print(remove_outliers.pulse.median())
```

```python
print(df.pulse.mean() - remove_outliers.pulse.mean()) 
#среднее арифимитическое изменилась
```

```python
print(df.pulse.median() - remove_outliers.pulse.median())
```

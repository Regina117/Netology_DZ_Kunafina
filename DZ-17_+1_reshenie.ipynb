```python
import pandas as pd
```

```python
df = pd.read_csv('horse_data.csv', header=None, na_values='?')
df = df[[1, 3, 8, 12, 17, 19, 20, 22]]
df.columns = ['Age',
              'rectal temperature °C',
              'mucous membranes', 
              'abdominal distension',
              'abdomen',
              'total protein',
              'abdominocentesis appearance',
              'outcome']
```

Age: Возраст

      1 = взрослая лошадь
      2 = Молодой (<6 месяцев)

rectal temperature °C : ректальная температура

      - линейный
      - в градусах Цельсия.
      - Повышенная температура может возникнуть из-за инфекции.
      - возможно снижение температуры при позднем шоке
      - нормальная температура 37,8
      - этот параметр обычно меняется по мере развития проблемы.
           например. может начаться нормально, а затем повыситься из-за
               поражение, возвращаясь в нормальный диапазон, как
               лошадь в шоке

mucous membranes : слизистые оболочки

      - субъективное измерение цвета
      - возможные значения:
           1 = нормальный розовый
           2 = ярко-розовый
           3 = бледно-розовый
           4 = бледно-синюшный
           5 = ярко-красный / инжектированный
           6 = темно-синюшный
      - 1 и 2, вероятно, указывают на нормальное или слегка повышенное
        обращение
      - 3 могут возникнуть при раннем шоке
      - 4 и 6 указывают на серьезное нарушение кровообращения
      - 5 больше указывает на сепсис

abdominal distension : вздутие живота

      - ВАЖНЫЙ параметр.
      - возможные значения
           1 = нет
           2 = легкий
           3 = умеренный
           4 = тяжелый
      - животное с вздутием живота может причинить боль и
        имеют сниженную перистальтику кишечника.
      - лошади с сильным вздутием живота, вероятно, потребуется
        Хирургия просто сними давление

abdomen : живот

      - возможные значения
           1 = нормальный
           2 = другое
           3 = твердый кал в толстой кишке
           4 = растянутый тонкий кишечник
           5 = растянутый толстый кишечник
      - 3, вероятно, препятствие, вызванное механическим воздействием
        и обычно лечится с медицинской точки зрения
      - 4 и 5 указывают на хирургическое поражение 

total protein : общий белок

      - линейный
      - нормальные значения лежат в диапазоне 6-7,5 (г / дл)
      - чем выше значение, тем сильнее обезвоживание 

abdominocentesis appearance : внешний вид брюшной полости

      - в брюшную полость лошади вводят иглу и получают жидкость из
        брюшная полость
      - возможные значения:
           1 = ясно
           2 = облачно
           3 = серозно-кровавый
      - нормальная жидкость прозрачная, в то время как мутная или серозно-кровавая указывает на
        скомпрометированный кишечник 

outcome : результат

      - что в итоге случилось с лошадью?
      - возможные значения:
           1 = жил
           2 = умер
           3 = был усыплен

# Задание 2.
Работа с выбросами В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

Рассмотрим показатель Pulse. при расчете около 10% удалилось, что показало отклонение среднее арифметическое изменилось на 1,65, это примерно на 2,5% от изначальных показателей. Думаю, что это отклонение при удалении выбросов допустимое при max = 184.0 и min = 30.0

```python
q1 = df['rectal temperature °C'].quantile(0.25)
q3 = df['rectal temperature °C'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['rectal temperature °C'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal temperature °C')
remove_outliers
```

```python

print(remove_outliers['rectal temperature °C'].mean())
print(remove_outliers['rectal temperature °C'].median())
```

```python
print(df['rectal temperature °C'].mean() - remove_outliers['rectal temperature °C'].mean())
```

```python
print(df['rectal temperature °C'].median() - remove_outliers['rectal temperature °C'].median())
```

По темперратуре выбросы мало искажают данные, наличе выбросов может говорить о других заболеваниях не связанные с коликами

Работа с пропусками Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

Выбраны были данные по pulse, но это не очень хороший пример все-таки, так как mode = 48.0, а mean = 64.0 и median = 71.9. при этом средний возрат mean = 2.6, а mean = 1.0 и median = 1.0. Когда максимальный возрат 9.0, а пульс 184.0. Очень сильное отклонение и нужно делить лошадей по возрасту и проводить оценку внутри возрастной группы.

```python
for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

```python
fill_df=df
fill_df['rectal temperature °C'].fillna(df['rectal temperature °C'].median(), inplace=True)
fill_df['mucous membranes'].fillna(1, inplace=True)
fill_df['abdominal distension'].fillna(3, inplace=True)
fill_df['total protein'].fillna(df['total protein'].median(), inplace=True)
fill_df['abdomen'].fillna(df['abdomen'].mode()[0], inplace=True)
fill_df['abdominocentesis appearance'].fillna(df['abdominocentesis appearance'].mode()[0], inplace=True)
fill_df['outcome'].fillna(1, inplace=True)
```

```python
for col in fill_df.columns:
    pct_missing = fill_df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

Вздутие живота: чаще всего было вздутие живота, проявлялось в разной стапени, чаще описывалось, как умеренное.
